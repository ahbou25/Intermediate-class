```python
#Working on the missing values Using (fillna) to fill up empty spaces in the columns
#Columns with the missinng values are float datatype filling them up with the average(mean)

```

```python
# Fill missing  for 'texture_mean' column using the mean
a = df_cancer['texture_mean'].mean()
df_cancer['texture_mean'].fillna(a, inplace=True)
```

```python
#Using a function fillnull to check the datatype of the column and fill the null values with its mean
def fillnull(column):
    if column.dtype is np.dtype(float):
        return column.fillna(column.mean())
    elif column.dtype is np.dtype(int):
        return column.fillna(column.mean())
    else:
        return column
```

```python
#calling the function with th dataframe
df_cancer = df_cancer.apply(fillnull)
```

```python
#Checking if the missing values have been filled 
df_cancer.isna().sum()
```

```python
#check for duplicated values 

df_cancer[df_cancer.duplicated()] #There are five rows and 32 columns with missing values
```

```python
df_cancer.dropna(inplace= True)
```
